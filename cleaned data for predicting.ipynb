```python
def convert(x):
    # Check if the value contains a range (i.e., contains a hyphen '-')
    if ' - ' in x:
        tokens = x.split(" - ")  # Split the range by ' - '
        try:
            # Convert both tokens to floats and calculate the average
            low = float(tokens[0].strip())
            high = float(tokens[1].strip())
            return (low + high) / 2  # Return the average
        except ValueError:
            return None  # Return None if conversion fails (invalid range)
    
    try:
        # If it's a single value (not a range), return the float value directly
        return float(x)
    except ValueError:
        return None  # Return None if conversion fails (invalid value)
```

```python
df["total_sqft"]=df["total_sqft"].apply(convert)
```

```python
df=df.dropna()
```

```python
df["price_per_sqt"]=df["price"] * 100000 / df["total_sqft"].astype(int)
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df["total_sqft"].quantile(0.25)
Q3 = df["total_sqft"].quantile(0.75)

# Calculate IQR (Interquartile Range)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier removal
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove rows where any value is outside the lower and upper bounds
outliers_total = df[(df["total_sqft"] >= lower_bound) & (df["total_sqft"] <= upper_bound)]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize the RandomForestRegressor model
rf_clf = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model with the training data
rf_clf.fit(X_train, Y_train)

# Model performance on test data
rf_score = rf_clf.score(X_test, Y_test)  # R^2 score on test data
print("Random Forest model R² score on test data:", rf_score)

# Making predictions on the test data
Y_pred = rf_clf.predict(X_test)

# Evaluating model performance
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")
```

```python
import joblib

# Save the trained model to a file
joblib.dump(rf_clf, 'random_forest_model.pkl')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}

grid_search = GridSearchCV(estimator=rf_clf, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, Y_train)

print("Best parameters found: ", grid_search.best_params_)
print("Best cross-validation score: ", grid_search.best_score_)
```

```python
# Evaluate the best model on the test set
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, Y_test)  # For regression tasks
print("Test set score: ", test_score)
```

```python
# For Random Forest Classifier
import matplotlib.pyplot as plt

feature_importances = best_model.feature_importances_
# Plot the feature importances
plt.barh(X_train.columns, feature_importances)
plt.xlabel("Feature Importance")
plt.ylabel("Feature")
plt.title("Feature Importance Plot")
plt.show()
```
